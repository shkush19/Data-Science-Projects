__OpenCV library to capture video from a webcam, convert each frame to grayscale, and display it in a window.__


```python
import cv2##Importing libraries for computer vision

cap = cv2.VideoCapture(0)##captures video from the default camera
##index(0) means , if you multiple camera installed then accordingly changed the index value

width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)##set the width of the captured frame
height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)##set the height of the captured frame

while True:##we used loop as camera will continously captures the video untill interrupted by the user
    ret, frame = cap.read()##successfully read the video
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)##convert the video frame to gray

    cv2.imshow('frame', gray)##for visualizing

    if cv2.waitKey(1) & 0xFF == ord('q'):##user can interrupt or loop will break the video capture by pressing 'q' from keyboard
        break

cap.release()##release the camera resource
cv2.destroyAllWindows()##closes the camera window
```
