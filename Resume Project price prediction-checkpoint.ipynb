<h3><font color="Black"><b>About the data--</b></font></h3>

> The dataset represents weekly 2018 retail scan data for National retail volume (units) and price. Retail scan data comes directly from retailers’ cash registers based on actual retail sales of Grapes. Starting in 2013, the table below reflects an expanded, multi-outlet retail data set. Multi-outlet reporting includes an aggregation of the following channels: grocery, mass, club, drug, dollar and military. The Average Price (of grapes) in the table reflects a per unit (per grapes) cost, even when multiple units (grapes) are sold in bags. The Product Lookup codes (PLU’s) in the table are only for grapes. Other varieties of grapes are not included in this table.

Some relevant columns in the dataset:

* Date - The date of the observation
* AveragePrice - the average price of a single bunch of Grapes
* type - conventional or organic
* year - the year
* Region - the city or region of the observation
* Total Volume - Total number of grapes sold
* 4046 - Total number of Grapes with PLU 4046 sold--PLU1
* 4225 - Total number of Grapes with PLU 4225 sold--PLU2
* 4770 - Total number of Grapes with PLU 4770 sold--PLU3

```python
# Creating dummy variables
dummies = pd.get_dummies(copied_grapes_data1[['year','region','Month']],drop_first=True)
df_dummies = pd.concat([copied_grapes_data1[['Total Volume', '4046', '4225', '4770', 'Total Bags',
       'Small Bags', 'Large Bags', 'XLarge Bags', 'type']],dummies],axis=1)
target = copied_grapes_data1['AveragePrice']
```

```python
# Splitting data into training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_dummies,target,test_size=0.30)
```

```python
# Standardizing the data
cols_to_std = ['Total Volume', '4046', '4225', '4770', 'Total Bags', 'Small Bags','Large Bags', 'XLarge Bags']
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train[cols_to_std])
X_train[cols_to_std] = scaler.transform(X_train[cols_to_std])
X_test[cols_to_std] = scaler.transform(X_test[cols_to_std])
```

```python
#importing ML models from scikit-learn
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
#to save time all models can be applied once using for loop
regressors = {
    'Linear Regression' : LinearRegression(),
    'Decision Tree' : DecisionTreeRegressor(),
    'Random Forest' : RandomForestRegressor(),
    'Support Vector Machines' : SVR(gamma=1),
    'K-nearest Neighbors' : KNeighborsRegressor(n_neighbors=1),
    'XGBoost' : XGBRegressor()
}
results=pd.DataFrame(columns=['MAE','MSE','R2-score'])
for method,func in regressors.items():
    model = func.fit(X_train,y_train)
    pred = model.predict(X_test)
    results.loc[method]= [np.round(mean_absolute_error(y_test,pred),3),
                          np.round(mean_squared_error(y_test,pred),3),
                          np.round(r2_score(y_test,pred),3)
                         ]
```

```python
##Q and A

def get_average(Ana1, column_name):
    """
    Description: This function returns the average value of the column.

    Arguments:
        Ana1: the DataFrame.
        column_name: the selected column.
    Returns:
        column's average.
    """
    column_sum = Ana1[column_name].sum()
    num_values = Ana1[column_name].count()
    return column_sum / num_values



result = get_average(Ana1, 'Total Bags')
print("Average:", result)

average_total_bags = get_average(Ana1, 'Total Bags')
print("Average Total Bags:", average_total_bags)
```

```python
def get_average_between_two_columns(Ana1, column1, column2):
    """
    Description: This function calculates the average between two columns in the dataset.

    Arguments:
        Ana1: the DataFrame.
        column1: the first column.
        column2: the second column.
    Returns:
        Sorted data for relation between column1 and column2.
    """
    unique_values = Ana1[column1].unique()
    averages = []

    for value in unique_values:
        subset = Ana1[Ana1[column1] == value]
        column2_average = subset[column2].mean()
        averages.append(column2_average)

    sorted_data = pd.DataFrame({column1: unique_values, column2: averages})
    sorted_data = sorted_data.sort_values(by=column2, ascending=False)

    return sorted_data

result = get_average_between_two_columns(Ana1, 'Total Bags', 'Small Bags')
print("Sorted data for relation between 'Total Bags' and 'Small Bags':")
print(result)
```

```python
def plot(data,xlabel,ylabel):
    """
    Description: This function to draw a barplot

    Arguments:
        data: the DataFrame. 
        xlabel: the label of the first column. 
        ylabel: the label of the second column.
    Returns:
        None
    """
        
    plt.figure(figsize=(15,5))
    ax=sns.barplot(x=data.column1,y=data.column2,palette='rocket')
    plt.xticks(rotation=90)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(('Average '+ylabel+' of Grapes According to '+xlabel));
```
