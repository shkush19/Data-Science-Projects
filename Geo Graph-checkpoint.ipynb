```python
import plotly.graph_objects as go
```

```python
data = {'lon': [20, 40, 60, 80],'lat': [77, 60, 70, 80],'value': [100, 200, 300, 400]}
```

```python
df = pd.DataFrame(data)
```

```python
# Plotting the Scattergeo map

fig = go.Figure(go.Scattergeo(
lon = df['lon'], # specify your longitude column
lat = df['lat'], # specify your latitude column
mode = 'markers',
marker = dict(
    size = 10,
    opacity = 0.8,
    reversescale = True,
    autocolorscale = False,
    symbol = 'circle',
    line = dict(width=1, color='rgba(102, 102, 102)'), colorscale = 'viridis', cmin = 0,
    color = df['value'],
    cmax = df['value'].max(), 
    colorbar_title="Colorbar Title"
)
))

fig.update_layout(
    title = 'Scattergeo Plot',
    geo=dict(scope='world',showland=True))

fig.show()
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Sample data
data = {
    'City': ['New York', 'Los Angeles', 'Chicago'],
    'Population': [8550405, 3971883, 2720546],
    'Latitude': [40.7128, 34.0522, 41.8781],
    'Longitude': [-74.0060, -118.2437, -87.6298]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a GeoDataFrame from DataFrame
geometry = [Point(xy) for xy in zip(df['Longitude'], df['Latitude'])]
gdf = gpd.GeoDataFrame(df, geometry=geometry)

# Print the GeoDataFrame
print(gdf)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Sample data
data = {
    'City': ['New York', 'Los Angeles', 'Chicago'],
    'Population': [8550405, 3971883, 2720546],
    'Latitude': [40.7128, 34.0522, 41.8781],
    'Longitude': [-74.0060, -118.2437, -87.6298]
}

# Create a GeoDataFrame from the sample data
gdf = gpd.GeoDataFrame(data, geometry=gpd.points_from_xy(data['Longitude'], data['Latitude']))

# Load world map data using geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Plot the map
fig, ax = plt.subplots(figsize=(10, 6))
world.plot(ax=ax, color='lightgray', edgecolor='black')
gdf.plot(ax=ax, marker='o', color='red', markersize=gdf['Population'] / 100000, alpha=0.5)

# Add title and labels
plt.title('Cities Population')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
data = {
    'City': ['New York', 'Los Angeles', 'Chicago'],
    'Population': [8550405, 3971883, 2720546]
}

# Extract city names and populations
cities = data['City']
populations = data['Population']

# Plot a pie chart
plt.figure(figsize=(8, 6))
plt.pie(populations, labels=cities, autopct='%1.1f%%', startangle=140)
plt.title('Population Distribution Among Cities')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the plot
plt.show()
```
