```python
from IPython.display import Audio
import pyttsx3

text = '''welcome to naresh technology datascience programme under prakash senapati guidance. 
        training will help practicle exposure to boost up technical skill 
        and increase learning for coding skills. 
        we conduct this programme for both non-technical & technical learners. Thank you.'''

audio = pyttsx3.init()
audio.setProperty('rate', 150)
audio.setProperty('volume', 0.8)

# Change the voice to a male voice
voice = audio.getProperty('voices')
audio.setProperty('voice', voice[0].id)  # Use voice[1].id for a female voice

# Convert text to speech
audio.save_to_file(text, 'test_male_Voice.mp3')
audio.runAndWait()

# Display the audio in Jupyter Notebook
Audio('test_male_Voice.mp3', autoplay=True)
```

```python
from IPython.display import Audio
import pyttsx3
from pydub import AudioSegment

# Define the text for male and female voices
male_text = "Hello, My name is Navneet, I am your male voice assistant."
female_text = "Hello, My name is Shikha, I am your female voice assistant."

# Initialize pyttsx3 instances for male and female voices
male_audio = pyttsx3.init()
female_audio = pyttsx3.init()

# Set properties for the male voice
male_voice = male_audio.getProperty('voices')[0]
male_audio.setProperty('voice', male_voice.id)
male_audio.save_to_file(male_text, "male_audio.wav")
male_audio.runAndWait()

# Set properties for the female voice
female_voice = female_audio.getProperty('voices')[1]
female_audio.setProperty('voice', female_voice.id)
female_audio.save_to_file(female_text, "female_audio.wav")
female_audio.runAndWait()

# Load male and female audio files
male_audio = AudioSegment.from_file("male_audio.wav")
female_audio = AudioSegment.from_file("female_audio.wav")

# Concatenate the audio files
concatenated_audio = male_audio + female_audio

# Export the concatenated audio file
concatenated_audio.export("concatenated_audio.wav", format="wav")

# Display the concatenated audio file in Jupyter Notebook
Audio("concatenated_audio.wav", rate=22050, autoplay=True)
```
