```python
clean_d['Rawmaterial']=clean_d['Rawmaterial'].fillna(np.mean(pan.to_numeric(clean_d['Rawmaterial'])))

##Filling missing values by mean strategy as this is the numeric data
```

```python
clean_d['Production']=clean_d['Production'].fillna(np.mean(pan.to_numeric(clean_d['Production'])))

##Filling missing values by mean strategy as this is the numeric data
```

```python
##Filling missing Categorical Data by Mode Strategy

clean_d
```

```python
clean_d['Location']
```

```python
clean_d['Location']=clean_d['Location'].fillna(clean_d['Location'].mode()[0])
```
